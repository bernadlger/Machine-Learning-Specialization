Goal: Implement the model ùëìùë§,ùëè for linear regression with one variable.

```python
import numpy as np                     # NumPy, a popular library for scientific computing
import matplotlib.pyplot as plt        # Matplotlib, a popular library for plotting data, also the module "pyplot" is the most used module in this lib.

# x_train is the input variable (size in 1000 square feet)
# y_train is the target (price in 1000s of dollars)

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f" x_train = {x_train}")
print(f" y_train = {y_train}")
```

```python
# m is the number of training examples
print( f" x_train.shape: {x_train.shape}")   #burada "shape" NumPy dizisinin boyutunu d√∂nd√ºr√ºr
m = x_train.shape[0]                         # x_train.shape[0], dizinin ilk boyutunun (satƒ±r sayƒ±sƒ±nƒ±n) eleman sayƒ±sƒ±nƒ± yani eƒüitim sayƒ±sƒ±nƒ± verir.
print ( f" Number of training examples is : {m}") 
```

```python
 i = 1
 x_i = x_train[i]    # We will use (x^i, y^i) do denote the i^th training example. 
 y_i = y_train[i]    # Yani dizinin i√ßine indeks vasƒ±tasƒ±yla eri≈üip eƒüitim √∂rneklerimizi g√∂r√ºyoruz. (x (0) , y (0)) = (1.0, 300.0) gibi.
 print(f"(x^{i} , y^{i}) = ({x_i}, {y_i})")
```

 Plotting the Data (Veilerin √áizilmesi): We can plot these two points using the scatter() function in the matplotlib library.

```python
# Plot the data points
plt.scatter( x_train, y_train, marker = 'x' , c= 'r') # "marker" noktalara ≈üekil verirken, "c" ile renk belirleniyor.
# Set the title
plt.title("Housing Prices") 
# Set the y-axis label
plt.ylabel("Price (in 1000s of dollars)")
# Set the x-axis label
plt.xlabel("Size (1000 sqft)")
plt.show()
```

The model function for  linear regression (which is a function that maps from x to y) is represented as:
f(x) = wx^(i) + b     #Aslƒ±nda doƒürunun denklemi :)
The formula above is how you can represent straight lines ( which needed for linear regression)

```python
w = 200
b = 100
print(f" w : {w}")
print(f" b : {b}")
```

Now we're going to construct a function for model output. We'll use for loop to calculate function.

```python
def compute_model_output ( x,w,b): # Fonksiyon tanƒ±mladƒ±k
    m = x.shape[0] # x dizisinin satƒ±r sayƒ±sƒ±nƒ± aldƒ±k √ß√ºnk√º bu ifade dizinin ilk boyutundaki(satƒ±r) eleman sayƒ±sƒ±nƒ± d√∂nd√ºr√ºyor.
    f_wb = np.zeros(m) # m uzunluƒüunda sƒ±fƒ±rlardan olu≈üan bir numpy dizisi olu≈üturduk. Bu dizi, tahmin deƒüerlerini (f_wb) saklamak i√ßin kullanƒ±lacak
    for i in range(m): # 0'dan m'e kadar (m dahil deƒüil) sayƒ±larƒ± √ºretir ve i her iterasyonda bu deƒüerleri alƒ±r.
        f_wb[i] = w * x[i] + b

    return f_wb
```

Now let's call the compute_model_output function and plot the output. A≈üaƒüƒ±da yazacaƒüƒ±mƒ±z kodda kendi verdiƒüimiz w,b deƒüerleri ile √ßizdiƒüimiz grafiƒüin,  veri noktalarƒ± ile uyumlu olup olmadƒ±ƒüƒ±nƒ± test edip buna g√∂re w ve b deƒüerlerini d√ºzenleyeceƒüiz. Nihai form√ºl√ºm√ºz√º ( olmasƒ± gereken w,b deƒüerlerini) elde etmi≈ü olacaƒüƒ±z. 

```python
tmp_f_wb = compute_model_output ( x_train, w, b)

plt.plot ( x_train, tmp_f_wb, c = "b" ,  label= "Our prediction") # Plot our model prediction
plt.scatter ( x_train, y_train, marker= "x", c= "r", label= "Actual Values") #Plot actual values
plt.title ("House Prices")
plt.ylabel("Price (in 1000s of dollars)")
plt.xlabel("Size (1000 sqft)")
plt.legend() # Grafikteki √ßizgi/nokta etiketlerini √ßizer.
plt.show() # Grafiƒüi ekrana √ßizer ve g√∂sterir.
```

Ba≈ülangƒ±√ßta verdiƒüimiz w=100 b= 100 deƒüerleri noktalarla uyumlu olmayƒ±nca √ºzerinde oynama yaptƒ±k ve w=200, b=100 deƒüerlerinin
grafiƒüe uyumlu olduƒüunu g√∂rd√ºk. w ve b deƒüerlerini bulduƒüumuza g√∂re artƒ±k modelimiz hazƒ±r :). 

Now that we have a model, we can use it to make our original prediction!

```python
w = 200
b = 100
x_i = 1.67
cost = w * x_i + b
print(f"$ {cost : 0.2f} thousand dollars")
```

Linear regression builds a model which establishes a relationship between features and targets
In the example above, the feature was house size and the target was house price
For simple linear regression, the model has two parameters  ùë§ and b whose values are 'fit' using training data.
Once a model's parameters have been determined, the model can be used to make predictions on novel data.
